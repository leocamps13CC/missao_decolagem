```python
# Dados simulados de telemetria

temperatura_interna = 22
temperatura_externa = -10
integridade_estrutural = 1
nivel_energia = 85
pressao_tanque = 5
modulos_criticos = 1

# Verificação das condições de segurança

if (
    18 <= temperatura_interna <= 27
    and -20 <= temperatura_externa <= 50
    and integridade_estrutural == 1
    and nivel_energia >= 60
    and 4 <= pressao_tanque <= 6
    and modulos_criticos == 1
):
    print("PRONTO PARA DECOLAR")
else:
    print("DECOLAGEM ABORTADA") 
```

# Simulação de Verificação de Decolagem

Este notebook demonstra a execução do sistema de verificação de telemetria
de uma nave antes da decolagem.

O sistema analisa:

- Temperatura interna
- Temperatura externa
- Integridade estrutural
- Nível de energia
- Pressão dos tanques
- Status dos módulos críticos
